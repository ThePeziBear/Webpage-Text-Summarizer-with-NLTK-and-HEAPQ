```python
def get_response(url):
    hdr = {'User-Agent':'Mozilla/5.4'}
    while True:
        try:
            with time_limit(100):
                content=requests.get(url,headers=hdr).content
                break
        except Exception as e:
            print('Request Timeout')
            raise e
        pass
    return content
```

```python
def get_soup(url):
    response = get_response(url)
    soup = bs(response,'lxml-xml')
    return soup
```
